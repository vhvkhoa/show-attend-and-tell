```python
import pickle
import json
import os
```

```python
def read_candidate(split):
    with open('data/%s/%s.candidate.captions.json' % (split, split), 'r') as f:
        return json.load(f)
def read_idx_to_id(split):
    with open('data/%s/%s.idxs_to_image_ids.pkl' % (split, split), 'r') as f:
        return pickle.load(f)
def read_image_names(split):
    with open('data/%s/%s.annotations.pkl' % (split, split), 'r') as f:
        return pickle.load(f)['file_name']    
def read_original_captions(split):
    with open('data/annotations/captions_%s2017.json' % (split), 'r') as f:
        ann = json.load(f)
    return ann['annotations']
def save_json(split, name, obj):
    with open('data/%s/%s.%s' % (split, split, name), 'w') as f:
        json.dump(obj, f)
```

```python
image_names = sorted(list(read_image_names('val')), key=lambda x: int(x.split('/')[-1][:-4].lstrip('0')))
candidate = read_candidate('val')
orig_caps = read_original_captions('val')
candidate = sorted(candidate, key=lambda x: x['image_id'])
orig_caps = sorted(orig_caps, key=lambda x: x['image_id'])
merged_orig_caps = {}
for cap in orig_caps:
    if cap['image_id'] not in merged_orig_caps.keys():
        merged_orig_caps[cap['image_id']] = [cap['caption']]
    else:
        merged_orig_caps[cap['image_id']].append(cap['caption'])
```

```python
for cap in candidate:
    img_id, caption = cap['image_id'], cap['caption']
    merged_orig_caps[img_id].append('Generated caption : ' + caption)
merged_orig_caps = sorted(merged_orig_caps.items(), key=lambda x: x[0])
```

```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from tqdm import tqdm

font = ImageFont.truetype("OpenSans-Bold.ttf", 20)
save_dir = 'image_val/'
for i, (image_id, caps) in tqdm(enumerate(merged_orig_caps)):
    if str(image_id) == image_names[i].split('/')[-1][:-4].lstrip('0'):
        file_name = image_names[i]
    else:
        print int(image_names[i].split('/')[-1][:-4].lstrip('0')), image_id
    im = Image.open(file_name)
    im = im.convert('RGB')
    background = Image.new(im.mode, (im.width, im.height+len(caps)*30), (256,)*3)
    background.paste(im)
    draw = ImageDraw.Draw(background)
    max_width = max([draw.textsize(cap, font)[0] for cap in caps])
    if max_width > im.size[0]:
        background = Image.new(im.mode, (max_width, im.height+len(caps)*30), (256,)*3)
        background.paste(im)
        draw = ImageDraw.Draw(background)
    for i, caption in enumerate(caps):
        draw.text((5, im.height + i*30), caption, font=font, fill='black')
    
    background.save(save_dir + file_name.split('/')[-1])
```
